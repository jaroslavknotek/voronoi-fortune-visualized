# Parabolae

```python
import matplotlib.pyplot as plt
import drawing
import geometry
from DataType import Point

focus_1 = Point(50,70)
focus_2 = Point(20,30)

fig, (axl,axr) = plt.subplots(1,2,figsize=(8,4))
plt.suptitle("Parabolae intersection")

drawing.draw_intersecting_parabolae_notebook(axl, focus_1,focus_2,  geometry.get_parabolae_intersection_euclid,drawing.draw_euclid_parabola , "euclidean" )
drawing.draw_intersecting_parabolae_notebook(axr, focus_1,focus_2,  geometry.get_parabolae_intersection_manhattan,drawing.draw_manhattan_parabola , "manhattan" )   
```

# Circles



```python
import numpy as np

fig, (axl,axr) = plt.subplots(1,2,figsize=(8,4))
plt.suptitle("Circle defined by three points")

points = [Point(x,y) for x,y in  [(300,10),(100,200),(450,350)]]
drawing.draw_circle_notebook(axl,points, drawing.draw_euclid_circle,geometry.circle_euclid,"euclidean")
drawing.draw_circle_notebook(axr,points, drawing.draw_circle_manhattan, geometry.circle_manhattan, "manhattan")
```
